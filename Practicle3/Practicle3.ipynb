```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'student_id': range(1, 11),
    'teacher_id': [101, 102, 103, 101, 104, 102, 103, 104, 101, 103],
    'rating': [4, 5, 3, 4, 2, 5, 4, 3, 5, 4],
    'beauty': [7, 8, 5, 6, 4, 7, 5, 6, 8, 5],
    'gender': ['M', 'F', 'M', 'F', 'F', 'M', 'F', 'M', 'F', 'M'],
    'tenure': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes'],
    'age': [35, 42, 38, 35, 45, 42, 38, 45, 35, 38],
    'course_level': ['lower', 'upper', 'lower', 'lower', 'upper', 'upper', 'lower', 'upper', 'lower', 'lower']
}

df = pd.DataFrame(data)
df
```

### Q1: Find duplicates

```python
duplicates = df[df.duplicated(subset='teacher_id', keep=False)]
print("Duplicate cases based on teacher_id:\n", duplicates)
```

### Age mean & std (all observations)

```python
mean_age_all = df['age'].mean()
std_age_all = df['age'].std()
print(f"All observations - Age Mean: {mean_age_all}, Std: {std_age_all}")
```

### Age mean & std (unique instructors, up to 94)

```python
df_unique = df.drop_duplicates(subset='teacher_id')
df_unique = df_unique.head(94)   # our dataset < 94 anyway
mean_age_unique = df_unique['age'].mean()
std_age_unique = df_unique['age'].std()
print(f"Unique instructors - Age Mean: {mean_age_unique}, Std: {std_age_unique}")
```

### Q2: Bar chart (lower vs upper courses)

```python
avg_rating_course = df.groupby('course_level')['rating'].mean()
avg_rating_course.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Average Teaching Ratings by Course Level")
plt.ylabel("Average Rating")
plt.xlabel("Course Level")
plt.show()
```

### Q3: Scatter (Age vs Teaching Evaluation)

```python
plt.scatter(df['age'], df['rating'], color='green')
plt.title("Age vs Teaching Evaluation")
plt.xlabel("Age")
plt.ylabel("Rating")
plt.show()
```

### Q4: Scatter (Age vs Rating by Gender)

```python
sns.scatterplot(data=df, x='age', y='rating', hue='gender', s=100)
plt.title("Age vs Teaching Evaluation by Gender")
plt.xlabel("Age")
plt.ylabel("Rating")
plt.show()
```

### Q5: Scatter (Age vs Rating by Gender + Tenure)

```python
sns.scatterplot(data=df, x='age', y='rating', hue='gender', style='tenure', s=120)
plt.title("Age vs Evaluation by Gender and Tenure")
plt.xlabel("Age")
plt.ylabel("Rating")
plt.show()
```
