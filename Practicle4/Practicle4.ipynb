#### Q1. Percentage of Visible Minorities Who Are Tenure Professors

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Create Dataset
# ----------------------------
data = {
    'Name': ['Asha', 'Rohan', 'Meena', 'Arjun', 'Neha', 'Rahul', 'Pooja', 'Kiran', 'Dev', 'Sara'],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'M', 'F'],
    'Age': [45, 39, 50, 34, 41, 29, 36, 53, 38, 42],
    'Tenure': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes'],
    'Visible_Minority': ['Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes'],
    'Evaluation': [4.1, 3.5, 4.6, 3.2, 4.8, 3.9, 3.7, 4.4, 3.3, 4.2]
}

df = pd.DataFrame(data)
print("Simple Student Rating Dataset:\n", df)
```

```python
# Count of visible minorities
visible_minority = df[df['Visible_Minority'] == 'Yes']
total_vm = len(visible_minority)

# Among them, those who are tenured
tenured_vm = visible_minority[visible_minority['Tenure'] == 'Yes']
percent_tenured_vm = (len(tenured_vm) / total_vm) * 100

print("Percentage of visible minorities who are tenured:", round(percent_tenured_vm, 2), "%")

# Small comment
if percent_tenured_vm < 50:
    print("→ Fewer visible minorities are tenured professors.")
else:
    print("→ Most visible minorities are tenured professors.")
```

### Q2. Does Average Age Differ by Tenure?

```python
# Group by Tenure and calculate mean & std
age_stats = df.groupby('Tenure')['Age'].agg(['mean', 'std'])
print("Average Age and SD by Tenure:\n", age_stats)
```

### Q3. Which Graph Works Better for Age Variable?

```python
# Histogram is better to show age distribution
plt.hist(df['Age'], bins=5, edgecolor='black')
plt.title("Distribution of Age of Professors")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Q4. Difference Between pyplot.bar and pyplot.barh + Graph for Gender

```python
gender_counts = df['Gender'].value_counts()

# Horizontal bar chart
plt.barh(gender_counts.index, gender_counts.values, color=['skyblue', 'pink'])
plt.title("Gender Distribution of Professors")
plt.xlabel("Count")
plt.ylabel("Gender")
plt.show()
```

### Q5. Median Evaluation Score for Tenured Professors

```python
median_eval_tenured = df[df['Tenure'] == 'Yes']['Evaluation'].median()
print("Median Evaluation Score for Tenured Professors:", median_eval_tenured)
```
